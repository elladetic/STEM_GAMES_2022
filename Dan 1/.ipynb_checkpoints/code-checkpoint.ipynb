```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#a dio;

def f(x):
    return (x **2)
    
def g(x):
    if x >= -1 and x <= 1:
        return 1
    else :
        return 0
    
k = np.linspace(-1,1,50)
sum = []

for i in range(50):
    sum_value = 0
    for j in range(i):
        sum_value = sum_value + f(k[j]) * g(k[i-j])
    sum.append(sum_value)
    
```

```python
plt.plot(k, sum)
plt.plot(k, 2 * k**2+2/3)
```
